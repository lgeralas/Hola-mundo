**Extracción de datos**

Filtros de datos: filtros que nos ayudan a buscar algun conjunto o datos en concreto, ya sean categóricos o númericos 

```python
#Importamos librerias requeridas
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io 
```

```python
#Convertir de dataframe 
df= pd.read_csv('file.csv')
df
```

**Filtros básicos para variables cualitativas**

```python
#Filtro por objeto 
filtro1=df[df['Entidad_Federativa']== 'Yucat n']
filtro1
```

```python
#Filtro por filas
#filtro2= df.iloc[300:302,: ] # filas consecutivas
filtro2=df.iloc[[1000,2000,3000],:] # filas no consecutivas[filas, columnas]
filtro2
```

```python
#filtro por columnas
#filtro3=df.iloc[:,4:7] #columnas consecutivas
filtro3=df.iloc[:,[1,3,5]]   #columnas no consecutivas
filtro3
```

```python
#combinación de filtros [filas,columnas]
filtromix=df.iloc[[1000,2,300],[0,4]]
filtromix
```

```python
#filtro filas vs columnas
#paso : leemos el contenido del archivo
from google.colab import files
uploaded=files.upload()
df=pd.read_csv(io.BytesIO(uploaded['file.csv']),index_col=2) #solo las categoricas pueden separarse por índice
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas
filtro4=df.loc[["Puebla","Durango"],["Poblacion_ocupada_en_la_industria_manufacturera"]]
filtro4
```

```python
#filtro cabecera 
filtro5=df.head()
filtro5
```

**Filtros básicos para variables cuantitativas**

```python
#filtro por comparación
filtro6=df[df['Periodo']>2012]
filtro6
```

**Filtros básicos con conectores lógicos**

```python
#filtro Y
filtro7=df[(df['Periodo']>2006)&(df['Sexo']=='Mujeres')]
filtro7
```
